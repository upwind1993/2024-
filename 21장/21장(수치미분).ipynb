21.7 파이썬 함수

```python
import numpy as np
def diffdata(x):
    n = len(x)
    xd = np.zeros(n-1)
    for i in range(n-1):
        xd[i] = x[i+1] - x[i]
    return xd
x = np.random.uniform(-1.,1.,100)
xdiff = diffdata(x)
import pylab
pylab.plot(x,c='k',ls=':',label='x data')
pylab.plot(xdiff,c='k',lw=0.5,label='diff data')
pylab.grid()
pylab.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 0.2 + 25.*x - 200.*x**2 + 675.*x**3 - 900.*x**4 + 400.*x**5

x = np.arange(0.,0.9,0.1)
y = f(x)
dx = np.diff(x)
dy = np.diff(y)
dydx = dy/dx
print(dydx)
n = len(x)
xmid = []
for i in range(n-1):
    xmid.append((x[i]+x[i+1])/2.)
xa = np.linspace(0.,0.8,100)
ya = 25. - 400.*xa + 2025.*xa**2 - 3600.*xa**3 + 2000.*xa**4

plt.figure(figsize=(8, 6))
plt.plot(xa, ya, label="Analytical derivative", color='black')  # 분석적 도함수 곡선
plt.plot(xmid, dydx, 'ko', label="Finite differences")  # 유한 차분 값
plt.xlabel("$x$")
plt.ylabel("$f'(x)$")
plt.legend()
plt.grid()
plt.title("Derivative Comparison")
plt.show()
```

```python
f = lambda x,y: y - x - 2.*x**2 -2.*x*y - y**2
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-2.,0.25,0.25)
y = np.arange(1.,3.25,0.25)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
fx,fy = np.gradient(Z,0.25)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.contour(X,Y,Z,colors='k')
ax.grid()
plt.quiver(X,Y,-fx,-fy,color='k',width=0.005)
```
